```python
#to print even numbers from 1-50
for i in range(1,51):
    if (i%2==0):
        print(f'{i:4d}',end="")
    else:
        pass
    i+=1
```

```python
#code to print multiplication table of a given number
n = int(input("enter the number whose table is required: "))
for i in range(1,11):
    a = n*i
    print(f'{n}x{i}={a}')
    i+=1
```

```python
a = []  # create list once outside the loop

n = int(input("Enter the length of your list: "))

for i in range(n):
    x = int(input("Enter integer element: "))
    a.append(x)

print("\nList items with their index:")
for i in range(len(a)):
    print(i, ":", a[i])
```

```python
#counting the number of vowels in "COMPUTER"
s = 'computer'

count=0
for i in range(len(s)):
    if (s[i] in "aeiou"):
        count+=1
print(count)        

```
