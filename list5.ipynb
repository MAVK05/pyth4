```python
#binary search on a list

def bin(lst,key):
    low = 0
    high = len(lst) -1

    while high >= low:
        mid = (high + low)//2
        if key == lst[mid]:
            print("Yes!!Found it",mid,"at the position is",lst[mid])
        elif key < lst[mid]:
            high = mid -1
        else:
            low = mid +1

#main
n = int(input("Enter the number of elements in the list: "))
l = []
for i in range(n):
    l.append(input())
print("The list to be searched is:",l)
n = int(input("enter the element to be searched"))
x = bin(l,n)
```
