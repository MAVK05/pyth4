```python
#default argument.py
def add(x , lst= None):
    if lst == None:
        lst = []
    if x not in lst:
        lst.append(x)    
    return lst

def main():
    list1 = add(1)
    print(list1)

    list2 = add(2)
    print(list2)

    list3 = add(3,[11,12,13,14])
    print(list3)

    list4 = add(4)
    print(list4) 

main()  
```

```python
def reverse(lst):
    result = [0] * len(lst)

    for i in range(0,len(lst)):
        result[i] = lst[len(lst) - 1 - i]

    return result  

def rev(lst):
    x = lst.reverse()
    return lst 


#main
l2 = [1,2,3,4,5,6]
print(reverse(l2))
print(rev(l2))
```

```python
#studying the binary search method 
arr = [10,20,30,40,50,60]
target = 10

low = 0
high = len(arr)-1


while low <= high:
    mid = (low + high)//2

    if arr[mid]== target:
       print("yes!found at index:",mid,"and the value to be found is:",arr[mid])
       break
    elif (arr[mid]<target):
       low = mid + 1
    else:
       high = mid -1
```
